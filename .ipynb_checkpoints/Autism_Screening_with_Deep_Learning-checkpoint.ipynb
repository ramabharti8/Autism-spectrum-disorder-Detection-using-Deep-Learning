```python
#importing libraries
import sys
import pandas as pd
import sklearn
import keras
```

```python
# import the dataset
file = 'Autism-Child-Data.txt'

# read the csv
data = pd.read_table(file, sep = ',', index_col = None)
```

```python
# print the shape of the DataFrame, so we can see how many examples we have
print(data.loc[:])
```

```python
# print out multiple patients at the same time
data.loc[:30]
```

```python
# print out a description of the dataframe
data.describe()
```

```python
# drop unwanted columns
data = data.drop(['result', 'age_desc'], axis=1)
```

```python
data.loc[:20]
```

```python
# create X and Y datasets for training
x = data.drop(columns=['Class'])
y = data['Class']
```

```python
x.loc[:20]
```

```python
# convert the data to categorical values - one-hot-encoded vectors
X = pd.get_dummies(x)
```

```python
# print the new categorical column labels
X.columns.values
```

```python
# print an example patient from the categorical data
X.loc[1]
```

```python
# convert the class data to categorical values - one-hot-encoded vectors
Y = pd.get_dummies(y)
```

```python
Y.iloc[:20]
```

```python
from sklearn import model_selection
# split the X and Y data into training and testing datasets
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size = 0.4)
```

```python
# build a neural network using Keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam

# define a function to build the keras model
def create_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=152, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, kernel_initializer='normal', activation='relu'))
    model.add(Dense(2, activation='sigmoid'))
    
    # compile model
    adam =  keras.optimizers.Adam(learning_rate=0.001)
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

model = create_model()

print(model.summary())
```

```python
# fit the model to the training data
model.fit(X_train, Y_train, epochs=50, batch_size=10, verbose = 1)
```

```python
# generate classification report using predictions for categorical model

from sklearn.metrics import classification_report, accuracy_score

predictions = (model.predict(X_test) > 0.5).astype("int32").ravel()[:192]
predictions
```

```python
# Check the unique values to see if they are binary
print(Y_test[['YES']].value_counts())

# Convert any non-zero values to 1
Y_test[['YES']] = Y_test[['YES']].apply(lambda x: x.apply(lambda y: 1 if y > 0 else 0))

# Check the unique values again to make sure they are binary
print(Y_test[['YES']].value_counts())

```

```python
# evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, Y_test, verbose=1)

# make predictions on the test data
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

# compute precision, recall, and f1-score
from sklearn.metrics import classification_report
print('Results for Categorical Model')
report = classification_report(Y_test, y_pred)
print(report)

print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```

```python
print('Results for Categorical Model')
print(accuracy_score(Y_test[['YES']], predictions))
print(classification_report(Y_test[['YES']], predictions))
```
