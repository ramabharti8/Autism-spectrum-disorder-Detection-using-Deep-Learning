```python
#import the required libraries
import numpy as np
import pandas as pd
import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
```

```python
# import the dataset
file = 'Autism-Child-Data.txt'

# read the csv
data = pd.read_table(file, sep = ',', index_col = None)
```

```python
# print the shape of the DataFrame, so we can see how many examples we have
print(data.loc[:])
```

```python
# print out multiple patients at the same time
data.loc[:30]
```

```python
# print out a description of the dataframe
data.describe()
```

```python
# drop unwanted columns
data = data.drop(['result', 'age_desc'], axis=1)

```

```python
data.loc[:20]
```

```python
# create X and Y datasets for training
x = data.drop(columns=['Class'])
y = data['Class']

```

```python
x.loc[:10]
```

```python
# convert the data to categorical values - one-hot-encoded vectors
X = pd.get_dummies(x)
#Y = pd.get_dummies(y)
```

```python
# print the new categorical column labels
X.columns.values
```

```python
# print an example patient from the categorical data
X.loc[1]
```

```python
# convert the class data to categorical values - one-hot-encoded vectors
Y = pd.get_dummies(y)
```

```python
Y.iloc[:10]
```

```python

# split the X and Y data into training and testing datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

```

```python
# preprocess the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


```

```python
# build a neural network using Keras
import keras
from keras.models import Sequential
from keras.layers import Dense

def create_model():
    # create model
    model = Sequential()
    model.add(Dense(12, input_dim=186, activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(2, activation='softmax'))

    # compile model
    adam = keras.optimizers.Adam(learning_rate=0.01)
    model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
    return model

model = create_model()
```

```python
# create the model
model = create_model()
```

```python
# define early stopping callback
early_stop = EarlyStopping(monitor='val_loss', patience=5)

```

```python
# fit the model to the training data
history=model.fit(X_train, Y_train, epochs=100, batch_size=10, verbose = 1, validation_split=0.2, callbacks=[early_stop])

```

```python
# generate classification report using predictions for categorical model

from sklearn.metrics import classification_report, accuracy_score

predictions = (model.predict(X_test) > 0.5).astype("int32").ravel()[:192]
predictions
```

```python
# evaluate the model on the test data
loss, accuracy = model.evaluate(X_test, Y_test, verbose=1)

# make predictions on the test data
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)


# compute precision, recall, and f1-score
from sklearn.metrics import classification_report
print('Results for Categorical Model')
report = classification_report(Y_test, y_pred)
print(report)


print('Test Loss:', loss)
print('Test Accuracy:', accuracy)
```

```python
import matplotlib.pyplot as plt

# Plot training and validation accuracy
plt.plot(history.history['accuracy'], color='red', lw=2, label='Train accuracy')
plt.plot(history.history['val_accuracy'], color='orange', lw=2, label='Validation accuracy')

# Plot training and validation loss
plt.plot(history.history['loss'], color='blue', lw=2, label='Train loss')
plt.plot(history.history['val_loss'], color='green', lw=2, label='Validation loss')

# Set the legend and axis labels
plt.legend(loc='best')
plt.xlabel('Epoch')
plt.ylabel('Metric value')
plt.title('Model performance')
plt.show()

```
